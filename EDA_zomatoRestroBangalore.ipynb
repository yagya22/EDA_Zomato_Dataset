```python
location_rating = df.groupby('location')['rate'].mean().sort_values(ascending=False)
```

```python
#this is also nan 
plt.figure(figsize=(10, 6))
location_rating.head(20).plot(kind='bar', color='blue')
plt.title('Top Locations by Average Rating')
plt.xlabel('Location')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
#top 5 restro types in bangalore 
top_5_rest_types =df['rest_type'].value_counts().nlargest(5)
plt.pie(top_5_rest_types, labels=top_5_rest_types.index, autopct='%1.2f%%', startangle=90)
plt.title('Top 5 Restaurant Types');
```

```python
outlets = df.groupby(['location', 'name']).size().reset_index(name='count')
top_outlets = outlets.sort_values(by='count', ascending=False)
top_outlets
```

```python
#now we wil find top 10 location of restro with its number of outleds in bangalore location
top10outlets=df['location'].value_counts().nlargest(10)
plt.pie(top10outlets, labels=top10outlets.index, autopct='%1.2f%%')
plt.title('Top 10 location with no of outlets percentage');
```

```python
top10_locations = df['location'].value_counts().nlargest(10)
```

```python
top10_data = df[df['location'].isin(top10_locations.index)]
location_restaurant_counts = top10_data.groupby(['location', 'name']).size().reset_index(name='count')
most_common_restaurant_per_location = location_restaurant_counts.sort_values(by=['location', 'count'], ascending=[True, False]).drop_duplicates('location')
```

```python
top10_labels = most_common_restaurant_per_location['location'] + ' - ' + most_common_restaurant_per_location['name']
top10_sizes = top10_locations.values
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top10_sizes, labels=top10_labels, autopct='%1.2f%%', startangle=90)
plt.title('Top 10 Locations with Most Common Restaurant Names and Outlet Count')
plt.axis('equal')
plt.show()
```
